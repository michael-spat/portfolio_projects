- выявить наиболее и наименее важные гипотезы, требующие первостепенного тестирования,
- провести и проанализировать A/B-тест,
- принять обоснованное решение по результатам тестирования.

## Приоритезация гипотез
На данном этапе нам необходимо выяснить, какие гипотезы подготовлены отделом маркетинга, а также оценить, насколько каждая конкретная гипотеза имеет первостепенную важность и требует срочного теста. Каждая гипотеза имеет свою оценку по параметрам `Reach`, `Impact`, `Confidence` и `Efforts`, что позволяет оценить их по фреймворкам `ICE` и `RICE`. По данным фреймворкам мы получим показатель, указывающим на степень срочности и важности конкретной гипотезы.

Для начала подготовим библиотеки и ознакомимся с актуальным для данного этапа датасетом `hypothesis.csv`.

```python
# импортируем необходимые для статистического анализа библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
```

Загрузим датасет в датафрейм `hypothesis`

```python
# загрузим датасет и загрузим его в датафрейм "hypothesis"
hypothesis = pd.read_csv('https://code.s3.yandex.net/datasets/hypothesis.csv')
```

```python
# выведем датафрейм
pd.set_option('max_colwidth', 200)
hypothesis
```

```python
# получим сводку о типах и структуре датафрейма
hypothesis.info()
```

Как мы видим, структура и типы датафрейма полностью соответствуют предполагаемым операциям с датафреймом: строки в столбце `Hypothesis` представлены типом `object`, а оценки параметров — типом `int64`. Однако мы наблюдаем неконвенциональное наименование столбцов с заглавной буквы. Исправим это.

```python
# приведем наименование столбцов к строчному написанию
hypothesis.columns = hypothesis.columns.str.lower()

# проверим изменения
hypothesis.columns
```

Теперь, когда датафрейм подготовлен к работе, ознакомимся с гипотезами.

```python
# выведем все гипотезы для ознакомления
counter = 0
for hypo in hypothesis['hypothesis']:
    display(f'Гиптоза № {counter}: ' + hypo)
    counter += 1
```

После того, как мы ознакомились, с гипотезами, мы можем приступить к их оценке по фремворкам `ICE` и `RICE`.

Освежим в памяти, как считают оценку по упомянутым фреймворкам и подготовим соответствующие функции для оперативного подсчета.

Фреймворк `ICE` позволяет оценить важность проверки гипотезы, основываясь на формуле:  

$\LARGE ICE = \frac{Impact \times Confidence}{Efforts} $  

На основе этой формулы мы можем подсчетать показатель `ICE` для каждой гипотезы.

```python
# высчитаем ICE для каждой гипотезы по формуле с округлением до 1 знака после запятой
hypothesis['ice'] = (hypothesis['impact']*hypothesis['confidence']/hypothesis['efforts']).round(1)
hypothesis.sort_values(by='ice', ascending=False)
```

В результате подсчета показателей `ICE` наиболее важным и срочным выступает гипотеза № 8 ("Запустить акцию, дающую скидку на товар в день рождения"), также высокие показатели наблюдаются у гипотез №№ 0 и 7: "Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей" и "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок".

Дополним показатели `ICE` оценками по фреймворку `RICE`, который высчитывается по следующей формуле:

$\LARGE RICE = \frac{Reach \times Impact \times Confidence}{Efforts} $  

По этой фурмуле подсчетаем показатель `RICE` для каждой гипотезы.

```python
# высчитаем RICE для каждой гипотезы по формуле с округлением до 1 знака после запятой
hypothesis['rice'] = (hypothesis['reach']*hypothesis['impact']*hypothesis['confidence']/hypothesis['efforts']).round(1)
hypothesis.sort_values(by='rice', ascending=False)
```

Лидером по важности и срочности в соответтсвии с показателем `RICE` выступает гипотеза № 7 "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок". Остальные гипотезы как минимум вдвое уступают лидеру.
Визуализируем полученные результаты.

```python
# задаим стиль
sns.set_style('darkgrid')

# визуализируем распределение оценок по двум фремворкам
hypothesis[['ice', 'rice']].sort_values(
    by = 'ice',
    ascending = False
).sort_values(
    by = 'rice', 
    ascending = False
).plot(
    kind = 'bar', 
    figsize = (15, 7),
    title = 'Показатели ICE и RICE у гипотез',
    xlabel = 'Гипотезы',
    ylabel = 'Оценка по фремворку',
    grid = True,
    rot = 0,
    yticks = np.arange(0, 120, step = 10)
)
plt.show()
```

Лидером по фреймворку `ICE` является гипотеза № 8, пока по фреймворку `RICE` лидер - гипотеза № 7. Разница результатов связана с добавлением параметра `reach` в формуле фреймворка `RICE`, где этот параметр выступает дополнительным множителем, а поскольку охват у гипотезы № 7 оценен максимально (10/10), то и оценка данной гипотезы меняется с 3-го на  1-ое место.  
Учитывая важность учета масштаба предпринимаемых изменений в продукте, предпочтительным явялется приоритезация гипотез по фреймворку `RICE`, поскольку гипотез в процессе работы над продуктом появляется достаточно много, для макимального эффекта требуется фокусировать первостпенные исследования на наиболее масшатбных изменениях.

## Анализ A/B-теста
На данном этапе мы уже провели A/B-тест и в результате получили 2 датасета: `orders` с заказами и `visitors` с количеством посещений на каждую день по группам.  
Загрузим датасеты в датафреймы и ознакомимся с данными.

```python
# загрузим датасет в датафрейм и ознакомимся с данными
orders = pd.read_csv('https://code.s3.yandex.net/datasets/orders.csv')
orders
```

```python
# загрузим датасет в датафрейм и ознакомимся с данными
visitors = pd.read_csv('https://code.s3.yandex.net/datasets/visitors.csv')
visitors
```

```python
# проверим структуру и типы в датафреймах
display(orders.info())
visitors.info()
```

В датафрейме `orders` мы наблюдаем неконвенциональное наименование столбцов, а также значение даты в формате `object`. Такое же несоответствие наблюдается и в датафрейме `visitors`. Исправим выявленные недочеты.

```python
# переименуем столбцы
orders = orders.rename(columns={'transactionId': 'transaction_id', 'visitorId': 'visitor_id'})

# проверим изменения
orders.columns
```

```python
# изменим даты в тип 'datetime'
orders['date'] = pd.to_datetime(orders['date'])
visitors['date'] = pd.to_datetime(visitors['date'])

# проверим структуру и типы в датафреймах
display(orders.info())
visitors.info()
```

После изменения типов также требуется проверить, нет ли в датафреймах дубликатов и все ли пользовтали относятся к отдельной группе теста.

```python
# проверим датафреймы на дубликаты
display(f'Выявлено {orders.duplicated().sum()} дубликатов в датафрейме orders')
f'Выявлено {visitors.duplicated().sum()} дубликатов в датафрейме orders'
```

Как мы видим дубликатов в представленных датафреймах не обнаружено.

После подготовки данных, проверим их на нормальность. Для этого расчитаем критерий Шапиро-Уилка.

```python
# разделим данные на сэмплы для датафрейма "orders"
sample_1a = orders[orders['group']=='A']['revenue']
sample_1b = orders[orders['group']=='B']['revenue']

# разделим данные на сэмплы для датафрейма "visitors"
sample_2a = visitors[visitors['group']=='A']['visitors']
sample_2b = visitors[visitors['group']=='B']['visitors']
```

```python
# задаем парметры проверки критерия - устанавливаем уровень стат.значимости в 5%
alpha = 0.05
# создаем список сэмплов на проверку
samples = {'sample_1a': sample_1a, 
           'sample_1b': sample_1b, 
           'sample_2a': sample_2a, 
           'sample_2b': sample_2b
          }

# циклом проверим все сэмплы
for name, sample in samples.items():
    # рассчитываем p-value
    results = st.shapiro(sample)
    # извлекаем p-value в отдельную переменную
    p_value = results[1]
    # создаем отчет с регулярными выражениями и вставляем названия сэмплов с p-value
    print(f'p-значение {name}: ', p_value)
    if p_value < alpha:
        print('Распределение не нормально\n')
    else:
        print('Распределение нормально\n')
```

Как мы видим, по критерию Шапиро-Уилка данные не распределены нормально, т.е. мы не можем использовать параметрические тесты с данными для получения достоверных результатов. В данном случае нам потребуется прибегнуть к непараметрическому тесту Манна-Уитни.  
Однако для начала изучим данные теста.

### График кумулятивной выручки по группам
Для начала посмотрим на кумулятивную выручку за август по группам теста.

```python
# подготовим датафрейм с заказами
dates_groups = orders[['date', 'group', 'transaction_id', 'visitor_id', 'revenue']]

# удаляем дубликаты по группам и дате
dates_groups = dates_groups.drop_duplicates(['date', 'group'])
dates_groups.head()
```

```python
# создаем новый датафрейм с агрегированными заказами
orders_aggregated = dates_groups.apply(
    lambda x: orders[
        np.logical_and(
            orders['date'] <= x['date'], orders['group'] == x['group']
        )
    ].agg(
        {
            'date': 'max',
            'group': 'max',
            'transaction_id': 'nunique',
            'visitor_id': 'nunique',
            'revenue': 'sum',
        }
    ),
    axis=1,
).sort_values(by=['date', 'group'])

# проверим изменения
orders_aggregated.head()
```

```python
# создаем новый датафрейм с агрегированными пользователями
visitors_aggregated = dates_groups.apply(
    lambda x: visitors[
        np.logical_and(
            visitors['date'] <= x['date'], visitors['group'] == x['group']
        )
    ].agg({'date': 'max', 'group': 'max', 'visitors': 'sum'}),
    axis=1,
).sort_values(by=['date', 'group'])

# проверим изменения
visitors_aggregated.head()
```

```python
# объединим агрегированные по заказам и пользователям датафреймы в один
cumulative_data = orders_aggregated.merge(
    visitors_aggregated, left_on=['date', 'group'], right_on=['date', 'group']
)

# переименуем столбцы объединенного датафрейма
cumulative_data.columns = [
    'date',
    'group',
    'orders',
    'buyers',
    'revenue',
    'visitors',
]

# проверим изменения
cumulative_data.head()
```

Мы получили кумулятивные данные по заказам и посетителям, однако полезно проверить, сохранность данных для теста. Проверим сохранился ли срок полученных данных после объединения.

```python
# сравним срок проведения теста
display(f'В датафрейме `orders` данные представлены за период: {orders["date"].min()} - {orders["date"].max()}')
f'В датафрейме `cumulative_data` данные представлены за период: {cumulative_data["date"].min()} - {cumulative_data["date"].max()}'
```

```python
# создадим срезы кумулятивных данных по группам теста
cumulative_revenue_a = cumulative_data[cumulative_data['group']=='A'][['date','revenue', 'orders']]
cumulative_revenue_b = cumulative_data[cumulative_data['group']=='B'][['date','revenue', 'orders']]

# визуализируем полученные срезы на графике кумулятивной выручки
plt.figure(figsize=(15, 10))
plt.plot(cumulative_revenue_a['date'], cumulative_revenue_a['revenue'], label='A')
plt.plot(cumulative_revenue_b['date'], cumulative_revenue_b['revenue'], label='B')
plt.legend()
plt.xlabel('Дни проведения теста')
plt.ylabel('Выручка с накоплением')
plt.title('График кумулятивной выручки по группам теста', {'fontsize': 14})
# перводим метки оси ординат из научного в стандартное отображение
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.show() 
```

Тренды обеих групп линейно растут на протяжении всего теста. Группы демонстрируют примерно схожие результаты выручки с небольшим перевесом в пользу группы В, однако на 20.08 произошел резкий скачок выручки у тестируемой группы, а затем характер роста продолжился с той же динамикой. Такой скачок сигнализирует о возможных выбросах в данных, что потребует дополнительной очистки последних.

### График кумулятивного среднего чека по группам
Дополним кумулятивный график выручки аналогичным графиком среднего чека по группам теста. Напомним, что средний чек представляет собой среднюю сумму, приходящуюся на один заказ. Следовательно средний чек высчитывается как отношение суммы выручки ко всем заказам:

$\LARGE \overline{bill} = \frac{\sum{revenue}}{\sum{orders}} $  

Для визуализации среденего чека потребуется по дням проведения теста в качестве оси абсцисс построить значения отношений выручки к заказам.

```python
plt.figure(figsize=(15, 10))
plt.plot(cumulative_revenue_a['date'], cumulative_revenue_a['revenue']/cumulative_revenue_a['orders'], label='A')
plt.plot(cumulative_revenue_b['date'], cumulative_revenue_b['revenue']/cumulative_revenue_b['orders'], label='B')
plt.legend()
plt.xlabel('Дни проведения теста')
plt.ylabel('Сумма среднего чека')
plt.title('График кумулятивного среднего чека по группам теста', {'fontsize': 14})
plt.show() 
```

График кумулятивного среднего чека к концу второй недели останавливается на уровне 7000, однако так же, как и с выручкой, мы наблюдаем резкий скачок тестируемой группы до уровня выше 10000. Однако дальнейший тренд идет на понижение, что дополнительно сигнализирует о выбросе в данных. Также следует отметить и падение среднего чека у контрольной группы под конец первой недели теста, что также может быть выбросом, хотя скачок и в пределах уровня. Если не учитывать потенциальные выбросы, мы наблюдаем преимущественно более высокий показатель среднего чека у тестируемой группы.  
Посмотрим, как изменения показателя среднего чека меняется у группы В относительно группы А.

### График относительного изменения кумулятивного среднего чека группы B к группе A
Построим график относительного изменения кумулятивного среднего чека группы В к группе А. Для этого нам потребуется выделить новый столбец со средним чеком в кумулятивных данных - `avg_bill`, а затем построить график, на котором по оси ординат будут представлены значения отношения среденго чека группы В к группе А за вычетом единицы:

$\LARGE \frac{\overline{bill_B}}{\overline{bill_A}}-1 $ 

```python
# добавим к датафрейму "cumulative_data" столбец с показателем среденего чека
cumulative_data['avg_bill'] = cumulative_data['revenue']/cumulative_data['orders']

# проверим изменения
cumulative_data.head()
```

```python
# создадим срезы по группам
cumulative_data_a = cumulative_data[cumulative_data['group']=='A']
cumulative_data_b = cumulative_data[cumulative_data['group']=='B']

# проверим изменения
display(cumulative_data_a.head())
cumulative_data_b.head()
```

```python
# соединим кумулятивные данные контрольной группы с тестовой
merged_cumulative_avg_bill = cumulative_data_a[['date','avg_bill']].merge(
    cumulative_data_b[['date','avg_bill']], 
    left_on='date', 
    right_on='date', 
    how='left', 
    suffixes=['_a', '_b']
)

# визуализируем данные и отводим отношение на ось ординат
plt.figure(figsize=(15, 10))
plt.plot(merged_cumulative_avg_bill['date'], 
         merged_cumulative_avg_bill['avg_bill_b']/merged_cumulative_avg_bill['avg_bill_a']-1, 
         label='Показатель относительного прироста кумулятивного среднего чека группы B относительно группы A')
plt.legend()

plt.axhline(y=0, color='black', linestyle='--')
plt.axhline(y=0.2, color='grey', linestyle='--')

plt.xlabel('Дни проведения теста')
plt.ylabel('Степень изменения среднего чека у группы B')
plt.title('График относительного изменения кумулятивного среднего чека группы B к группе A', {'fontsize': 14})
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)

plt.show()
```

Относительное изменение среднего чека группы В к группе А достаточно нестабильно: за первую неделю наблюдается скачок до 0,35, а затем на второй неделе резкое падение до отрицательных значений. Как и ожидалось, на 20-ый день отношение резко возрастает до 0,5, что отражает уже выявленный ранее выброс. Данные точно содеражт выбросы, требующие очистки.  
Введем вторую переменную - посетителей и рассмотрим, как заказы распределены по ним в группах теста.

### График кумулятивного среднего количества заказов на посетителя по группам

Добавим новый столбец `orders_per_visitor` с средним количеством заказов на посетителя в качестве отношения заказов к количеству посетителей, а затем сделаем срезы по группам.

### Cтатистическая значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным


Перед подсчетом статистической значимости требуется подготовить данные. Создадим для начала кумулятивные данные по каждой из групп, а затем объединим их в один датафрейм, на основе которого уже будем извлекать выборки для проверки гипотез.

```python
# создадим отдельные датафреймы c датой и количеством пользователей на каждую дату с фильтрацией по группе А
visitors_a_daily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitors_a_daily.columns = ['date', 'visitors_per_date_a']

# проверим результат
visitors_a_daily.head()
```

```python
# представим данные по пользователям в датафрейме по группе А в кумулятивном виде
visitors_a_cummulative = visitors_a_daily.apply(
    lambda x: visitors_a_daily[visitors_a_daily['date'] <= x['date']].agg(
        {'date': 'max', 'visitors_per_date_a': 'sum'}
    ),
    axis=1,
)
visitors_a_cummulative.columns = ['date', 'visitors_cummulative_a']

# проверим результат
visitors_a_cummulative.head()
```

```python
# создадим новый датафрейм группы А, где из исходного датафрейма "orders" на каждую дату приходится количество заказов и выручка
orders_a_daily = (
    orders[orders['group'] == 'A'][['date', 'transaction_id', 'visitor_id', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'})
)
orders_a_daily.columns = ['date', 'orders_per_date_a', 'revenue_per_date_a']

# проверим результат
orders_a_daily.head()
```

```python
# представим данные по заказам в датафрейме по группе А в кумулятивном виде
orders_a_cummulative = orders_a_daily.apply(
    lambda x: orders_a_daily[orders_a_daily['date'] <= x['date']].agg(
        {'date': 'max', 'orders_per_date_a': 'sum', 'revenue_per_date_a': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
orders_a_cummulative.columns = [
    'date',
    'orders_cummulative_a',
    'revenue_cummulative_a',
]

# проверим результат
orders_a_cummulative.head()
```

```python
# произведем те же операции для группы В
# # создадим отдельные датафреймы c датой и количеством пользователей на каждую дату с фильтрацией по группе B
visitors_b_daily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitors_b_daily.columns = ['date', 'visitors_per_date_b']

# представим данные по пользователям в датафрейме по группе В в кумулятивном виде
visitors_b_cummulative = visitors_b_daily.apply(
    lambda x: visitors_b_daily[visitors_b_daily['date'] <= x['date']].agg(
        {'date': 'max', 'visitors_per_date_b': 'sum'}
    ),
    axis=1,
)
visitors_b_cummulative.columns = ['date', 'visitors_cummulative_b']

# проверим результат
display(visitors_b_cummulative.head())

# создадим новый датафрейм группы В, где из исходного датафрейма "orders" на каждую дату приходится количество заказов и выручка 
orders_b_daily = (
    orders[orders['group'] == 'B'][['date', 'transaction_id', 'visitor_id', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'})
)
orders_b_daily.columns = ['date', 'orders_per_date_b', 'revenue_per_date_b']

# представим данные по заказам в датафрейме по группе В в кумулятивном виде
orders_b_cummulative = orders_b_daily.apply(
    lambda x: orders_b_daily[orders_b_daily['date'] <= x['date']].agg(
        {'date': 'max', 'orders_per_date_b': 'sum', 'revenue_per_date_b': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
orders_b_cummulative.columns = [
    'date',
    'ordersCummulativeB',
    'revenueCummulativeB',
]

# проверим результат
orders_b_cummulative.head()
```

Мы подготовили кумулятивные данные по каждой из групп, включая дату, количество посетителей, заказов и объем выручки с накоплением. Теперь мы объединяем данные в один датафрейм.

```python
# объединим данные в один датафрейм "data"
data = (
    orders_a_daily.merge(
        orders_b_daily, left_on='date', right_on='date', how='left'
    )
    .merge(orders_a_cummulative, left_on='date', right_on='date', how='left')
    .merge(orders_b_cummulative, left_on='date', right_on='date', how='left')
    .merge(visitors_a_daily, left_on='date', right_on='date', how='left')
    .merge(visitors_b_daily, left_on='date', right_on='date', how='left')
    .merge(visitors_a_cummulative, left_on='date', right_on='date', how='left')
    .merge(visitors_b_cummulative, left_on='date', right_on='date', how='left')
)

# проверим результат
data.head() 
```

Для подсчета статистической значимости разлий нам недостаточно полученной таблицы `data` - нам потребуется создать дополнительные переменные: `orders_by_users_a` и `orders_by_users_b`.

```python
# создаем переменную "orders_by_users_a" через группировку и аггрегацию числа уникальных заказов для группы А
orders_by_users_a = (
    orders[orders['group'] == 'A']
    .groupby('visitor_id', as_index=False)
    .agg({'transaction_id': pd.Series.nunique})
)
# переименуем столбцы
orders_by_users_a.columns = ['user_id', 'orders']

# проверим результат
display(orders_by_users_a.head())

# повторим операцию для группы В
orders_by_users_b = (
    orders[orders['group'] == 'B']
    .groupby('visitor_id', as_index=False)
    .agg({'transaction_id': pd.Series.nunique})
)
# переименуем столбцы
orders_by_users_b.columns = ['user_id', 'orders'] 

# проверим результат
orders_by_users_b.head()
```

Теперь на основе полученных данных создадим выборки на каждую группу: `sample_a` и `sample_b`. В каждой будет переменной пользователям из группы будет соответствовать количество заказов.

```python
# создадим выборку "sample_a"
sample_a = pd.concat(
    [orders_by_users_a['orders'],
     pd.Series(
         0, 
         index=np.arange(
             data['visitors_per_date_a'].sum() - len(orders_by_users_a['orders']
                                                 )
         ), 
         name='orders')
    ],
    axis=0
)

# проверим результат
sample_a.head()
```

```python
# создадим выборку "sample_b"
sample_b = pd.concat(
    [orders_by_users_b['orders'],
     pd.Series(
         0, 
         index=np.arange(
             data['visitors_per_date_b'].sum() - len(orders_by_users_b['orders']
                                                 )
         ), 
         name='orders')
    ],
    axis=0
)

# проверим результат
sample_b.head()
```

Теперь, когда данные готовы, мы можем проверять гипотезы. Как мы выяснили, располагаемые нами данные не распределены нормально, поэтому t-тест здесь не даст результатов, заслуживающих доверия. Для корректных результатов обратимся к критерию Манна-Уитни, чтобы выявить (или не обнаружить) статистическую значимость различий в среднем количестве заказов на посетителя 2-х групп теста.  
Соответственно, мы формулируем такие гипотезы:

$ \Large H_0 =$ ***различий в среднем количестве заказов на посетителя между группами A и B нет***  

$ \Large H_1 =$ ***различия в среднем количестве заказов на посетителя между группами A и B есть***

Для проверки гипотез мы уже располагаем необходимыми выборками:  
- группа А: `sample_a`
- группа B: `sample_b`

Уровень значимости установим в 5%, благодаря чему мощность теста будет составлять 95%.  
Теперь проведем U-тест.

```python
# устанавливаем уровень статистической значимости
alpha = 0.05

# сохраняем в отдельную переменную результаты теста Манна-Уитни, соответствующему методу выборки групп
results = st.mannwhitneyu(
    sample_a, 
    sample_b
)

# выведем результат и сравним p-value c заданным уровнем стат.значимости
print('Статистическая значимость (p-значение): ', results.pvalue.round(3))
print(f'Относительное различие: {((sample_b.mean() / sample_a.mean() - 1)*100).round(2)}%')
if results.pvalue < alpha:
    print('Отвергаем H0, поскольку разница присутствует и она статистически значима')
else:
    print('Придерживаемся H0: вывод о различии сделать нельзя')
```

Мы получили p-value больше уровня стат.значимости (0.017>0,05), а это значит, что различия в среденем количестве достаточны для  утверждения их статичестической значимости. Если ограничиваться только сырыми данными, то мы можем утверждать, что группа В полижительно отличается от группы А, т.е. эти группы демонстрируют разное покупательское поведение. Дополнительно отметим относительное различие между группами: 13.81%. Однако не стоит сразу же принимать результаты за чистую монету, т.к. мы уже выяснили, что данные имеют выбросы.  
Посмотрим, каковы ститстические отличия групп по среднему чеку.

### Статистическая значимость различий в среднем чеке заказа между группами по «сырым» данным
Повторим тест Манна-Уитни, но теперь выявим (или не обнаружим) статистическую занчимость различий в среднем чеке заказов между 2-мя группами теста.  
Соответственно, мы формулируем следующие гипотезы:

$ \Large H_0 =$ ***различий в среднем чеке между группами А и B нет***  

$ \Large H_1 =$ ***различия в среднем чеке между группами А и B есть***

Для проверки гипотез мы уже располагаем необходимыми выборками:  
- группа А: `orders[orders['group']=='A']['revenue']`
- группа B: `orders[orders['group']=='B']['revenue']`

Уровень значимости установим в 5%, благодаря чему мощность теста будет составлять 95%.  
Теперь проведем U-тест.

```python
# устанавливаем уровень статистической значимости
alpha = 0.05

# сохраняем в отдельную переменную результаты теста Манна-Уитни, соответствующему методу выборки групп
results = st.mannwhitneyu(
    orders[orders['group']=='A']['revenue'], 
    orders[orders['group']=='B']['revenue']
)

# выведем результат и сравним p-value c заданным уровнем стат.значимости
print('Статистическая значимость (p-значение): ', results.pvalue.round(3))
print(f'Относительное различие: {((orders[orders["group"]=="B"]["revenue"].mean() / orders[orders["group"]=="A"]["revenue"].mean() - 1)*100).round(2)}%')

if results.pvalue < alpha:
    print('Отвергаем H0, поскольку разница присутствует и она статистически значима')
else:
    print('Придерживаемся H0: вывод о различии сделать нельзя')
```

Мы получили p-value больше уровня значимости, значит предварительно мы можем придерживаться нулевой гипотезы о том, что группа B имеет такой же средний чек, что и у группы А. Такой вывод несколько противоречит высокому относительному различию в 25.87% и тому, что мы наблюдали на графиках - зачастую графики на последеней неделе стабилизируются с группой В, имеющей более высокие показатели на фоне группы А. Однако важно помнить, что данные не очищены, а показатель среднего чека крайне подвержен выбросам, поэтому наши выводы по двум тестам нам следует провести по уже очищенным данным.

### Очистка данных
Как мы выяснили, в параметрах количества заказов и среднего чека на пользователя имеются выбросы. Избавимся от них по следующим пороговым значениям:  

- для количества заказов на пользователя возьмем значение 1 (95-ый перцентиль),
- для среднего чека на пользователя возьмем пороговое значение 58 233.2 (99-ый перцентиль).

Обновим данные путем выявления крайних значений (пользователи, превышающие заданные перцентили), а затем отфильтруем выборки уже без выявленных пользователей.

```python
# найдем пользователей с большим количеством заказов
users_extra_orders = pd.concat(
    [
        orders_by_users_a[orders_by_users_a['orders'] > np.percentile(orders_per_users["transaction_id"], 95)]['user_id'],
        orders_by_users_b[orders_by_users_b['orders'] > np.percentile(orders_per_users["transaction_id"], 95)]['user_id'],
    ],
    axis=0,
)

# найдем пользователей с дорогими заказами
users_expensive_orders = orders[orders['revenue'] > np.percentile(orders["revenue"], 99)]['visitor_id']

# объединим полученные данные в единый список пользователей, от которых необходимо избавиться для очистки данных
abnormal_users = (
    pd.concat([users_extra_orders, users_expensive_orders], axis=0)
    .drop_duplicates()
    .sort_values()
)

# проверим результат и узнаем, сколько всего таких пользователей
display(abnormal_users.head(5))
f'Всего пользователей с экстремальными значениями: {abnormal_users.shape[0]}'
```

Мы получили список пользователей с экстремальными значениями - 59 пользователей. Теперь очистим наши выборки, убрав данные о выявленных пользователях.

```python
# очистим данные от крайних значений в выборке для группы А
filtered_sample_a = pd.concat(
    [orders_by_users_a[np.logical_not(orders_by_users_a['user_id'].isin(abnormal_users))]['orders'],
     pd.Series(0, index=np.arange(data['visitors_per_date_a'].sum() - len(orders_by_users_a['orders'])), name='orders')], axis=0,
)

# проверим резудьтат
display(filtered_sample_a.head())

filtered_sample_b = pd.concat(
    [orders_by_users_b[np.logical_not(orders_by_users_b['user_id'].isin(abnormal_users))]['orders'],
     pd.Series(0, index=np.arange(data['visitors_per_date_b'].sum() - len(orders_by_users_b['orders'])), name='orders')], axis=0,
)

# проверим резудьтат
filtered_sample_b.head()
```

### Cтатистическая значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным
 


Повторно сформулируем гипотезы:

$ \Large H_0 =$ ***различий в среднем количестве заказов на посетителя между группами A и B нет***  

$ \Large H_1 =$ ***различия в среднем количестве заказов на посетителя между группами A и B есть***

Для проверки гипотез используем очищенные выборки:  
- группа А: `filtered_sample_a`
- группа B: `filtered_sample_a`

Уровень значимости сохраним 5%.  
Теперь повторно проведем U-тест на очищенных выборках.

```python
# устанавливаем уровень статистической значимости
alpha = 0.05

# сохраняем в отдельную переменную результаты теста Манна-Уитни, соответствующему методу выборки групп
results = st.mannwhitneyu(
    filtered_sample_a, 
    filtered_sample_b
)

# выведем результат и сравним p-value c заданным уровнем стат.значимости
print('Статистическая значимость (p-значение): ', results.pvalue.round(3))
print(f'Относительное различие: {((filtered_sample_b.mean() / filtered_sample_a.mean() - 1)*100).round(2)}%')

if results.pvalue < alpha:
    print('Отвергаем H0, поскольку разница присутствует и она статистически значима')
else:
    print('Придерживаемся H0: вывод о различии сделать нельзя')
```

Избавившись от выбросов, мы наблюдаем иной результат проверки гипотезы о равенстве колиства заказов на посетителя в группах А и В: мы получаем p-value меньше уровня значимости (0.01<0.05), поэтому отвернаем нулевую гипотезу и принимаем альтернативную об их неравенстве. Если мы учтем все то, что демонстрировали графики, а также то, что относительное различие составляет ~17,5%, то мы можем с 95% увернностью утверждать, что у группы В количество заказов на посетителя больше, чем у посетителя из группы А.

Повторно сформулируем гипотезы для различий в среднем чеке:

$ \Large H_0 =$ ***различий в среднем чеке между группами А и B нет***  

$ \Large H_1 =$ ***различия в среднем чеке между группами А и B есть***

Будут использованы очищенные выборки для групп:
