## Описание проекта
Интернет-магазину «Стримчик» требуется выявить определяющие успешность игры закономерности. Для достижения этой цели принято решение воспользоваться данными, доступными из открытых источников, - историческая сводка о продажах игр, включающая оценки пользователей и экспертов, жанры и платформы. ## Описание данных
Предоставленный датасет `games.csv` содержит информацию об продажах видеоигр до 2016 года включительно. Датасет описывает игры по следующим признакам:
- название игры (столбец `Name`)
- платформа, на которой выпускалась игра (`Platform`)
- год релиза (`Year_of_Release`)

- возрастной рейтинг от ассоциации ESRB (`Rating`)

## 2. Предобработка данных
На данном этапе потребуется скорректировать названия столбцов, привести их данные к более подходящим типам, обработать выявленные и неявные пропуски, явные и неявные дубликаты, а также произвести обогащение данных путем добавления столбца с общими продажами.

### 2.1. Корректировка наименования столбцов
Как мы выяснили, имена столбцов начинаются с загланых букв, что противоречит общепринятой практике. Для этого воспользуемся методами `rename()` и `lower()`, а также обернем процедуру в цикл.

```python
# переименуем столбцы датафрейма
for name in data.columns:
    data = data.rename(columns={name:name.lower()})

# проверим результат
data.columns
```

Здесь же в дополнение приведем названия, платформы и жанры тайтлов к нижнему регистру для оптимизации агрегации в дальнейшем.

```python
# приведем избранные столбцы к нижнему регистру
columns_to_lower = ['name', 'platform', 'genre', 'rating']
for col in columns_to_lower:
    data[col] = data[col].str.lower()

data
```

### 2.2. Обработка пропусков
На предыдущем ознакомительном этапе мы уже выявили явные и неявные пропуски, однако следует учитывать и дубликаты как вариант уже представленного значения.

#### 2.2.1. Явные пропуски и дубликаты
Из столбцов с явными пропусками мы обнаружили `Name`, `Year_of_Release`, `Genre`, `Critic_Score`, `User_Score` и `Rating`. Рассмотрим эти столбцы подробнее.

```python
# выведем количество пропусков для всех столбцов
data.isna().sum()
```

Столбцы `name` и `genre` имеют по два пропуска, пока в остальных столбцах с пропусками объем намного больше, поэтому сосредоточимся сначала на малых пропусках.

```python
# посмотрим на 2 пропуска в столбцах 'name' и 'genre'
data[(data['name'].isna())|(data['genre'].isna())]
```

Как оказалось, два пропуска в столбцах `name` и `genre` - одни и те же пропуски, что объясняет корреляцию между ними на прошлом этапе. От них мы можем избавиться.

```python
# удалим явные пропуски в столбцах 'name' и 'genre'
data.dropna(inplace=True, subset=['name', 'genre'])
data.isna().sum()
```

От пропусков с годом выпуска мы избавимся, т.к. это менее 2% от датафрейма, а пропуски с оценками и рейтингом мы заменим на значения-заглушки. Такие значения имеют явно не правдоподобный характер, но в то же время они не будут путаться с реальными данными. В дальнейшем потребуется их отфильтровать, поэтому наиболее подходящим значением будет `-1`.

```python
# удалим пропуски в столбце "year_of_release"
data = data.dropna(subset=['year_of_release']).reset_index(drop=True)
data.isna().sum()
```

Остальные пропуски (кроме столбца `'rating'`) заполним значением-заглушкой `-1`.

```python
# заполним пропуски заглушкой "-1"
data[['critic_score', 'user_score']] = data[['critic_score', 'user_score']].fillna(-1)
data.isna().sum()
```

Явных пропусков в числовых столбцах теперь нет, а в столбце `'rating'` мы заменим пропуски на `'unknown'`. 

```python
# заполним пропуски заглушкой "-1"
data['rating'] = data['rating'].fillna('unknown')
data.isna().sum()
```

Помимо явных пропусков проверим датафрейм на наличие явных дубликатов.

```python
# проверим датафрейм на наличие явных дубликатов
data.duplicated().sum()
```

Явных дубликатов не обнаружено. Теперь обработаем неявные пропуски и дубликаты.

#### 2.2.2. Неявные пропуски и дубликаты
Как мы уже выяснили на этапе обзора данных, у нас есть следующие неявные пропуски и дубликаты:
- стобец `user_score` имеет значение `'tbd'`, которое в контексте данных то же самое, что и пропуск,
- стобец `rating` имеет значение `'RP'` (тоже пропуск) и `'K-A'`, которое равнозначно `'E'`.

Значения `'tbd'` и `'RP'` мы заменим на -1, а `'K-A'` заменим на `'E'`, но сначала проверим датафрейм на дубликаты по названию, году и платформе, т.к. одна и та же игра, но на разных платформах в разный год релиза является отдельной записью.

```python
# проверим наличие дубликатов постолбцам 'name', 'platform', 'year_of_release'
data.duplicated(subset=['name', 'platform', 'year_of_release']).sum()
```

В датафрейме всего 1 дубликат на отдельный релиз. Удалим его.

```python
# удалим выявленный дубликат
data = data.drop_duplicates(subset=['name', 'platform', 'year_of_release'], keep='first').reset_index(drop=True)

# проверим результат
data.duplicated(subset=['name', 'platform', 'year_of_release']).sum()
```

Теперь заменим выявленные неявные пропуски и неявные дубликаты на значения-заглушки.

```python
# заменим значения 'tbd' и 'rp' на -1
data = data.replace(
    to_replace = {'user_score': 'tbd', 'rating': 'rp'}, 
    value = {'user_score': '-1', 'rating': 'unknown'}
)

# заменим значения 'k-a' на 'e'
data = data.replace(
    to_replace = {'rating': 'k-a'}, 
    value = {'rating': 'e'}
)
```

Теперь, когда мы избавились от пропусков дубликатов, мы можем привести типы данных к подразумеваемому типу.

### 2.3. Преобразование типов данных
Как мы выяснили на предыдущем этапе всего в датафрейме используется 2 типа: `float64` и `object`, хотя предоставляемая информация явно не ограничивается имеющимися типами. Еще раз ознакомимся со столбцами:

```python
# выведем первые строки датафрейма
data.head(10)
```

Также пользовательские оценки представлены в строковом виде, хотя они явно должны быть представлены вещественным числом.
Скорректируем указанные столбцы и их типы данных.

```python
# изменим тип данных столбцов на int
data['year_of_release'] = data['year_of_release'].astype('int')
data['critic_score'] = data['critic_score'].astype('int')


# проверим типы
data[['year_of_release', 'critic_score']].dtypes
```

Тип данных изменен успешно. Теперь изменим пользовательские оценки в тип `float`.

```python
# изменим тип данных на flaot
data['user_score'] = pd.to_numeric(data['user_score'], errors='coerce')

# проверим тип данных
data['user_score'].dtypes
```

Данный этап позволит получить основную информацию о рынке видеоигр и спланировать стратегию дальнейшего развития интернет-магазина.

### 3.1. Сколько игр выпускалось за разные годы?
Для начала требуется выяснить динамику выпуска до 2016, чтобы получить представление о масштабах игрового рынка, а также о динамике его роста. Для начала выявим общее количество выпущенных игр за каждый год.

```python
# подсчитаем, сколько игр выпускалось в каждом году
games_number = data.pivot_table(
    index = 'year_of_release', 
    values = 'name',
    aggfunc = {'name':'count'}
)
games_number = games_number.reset_index(drop=False)
games_number
```

Прежде, чем анализировать результаты, следует отметить, что имеющиеся данные не отражают в полной мере каждый этап игрового рынка, поскольку сбор данных по данной сфере начался намного позже появления феномена видеоигр. По этой причине для ранних этапов индустрии следует учитывать неполноту картины.

Популярные платформы получают больше портов, а игры увеличивают число проданных копий с увеличением охвата платформ. Посмотрим на количество проданных копий у каждой платформы и выявим наиболее выдающиеся по своим продажам.

```python
# построим график без артефакта
leaders.plot(kind = 'line', 
             grid = True, 
             style = 'o-', 
             figsize = (20, 10), 
             xlabel = 'Год', 
             ylabel = 'Продажи (млн)', 
             title = 'Динамика продаж игр на популярных платформах до 2016 г.',
             xlim = [1993, 2017]
)
plt.show()
```

Полученный график продаж среди лидеров платформ мы также можем дополнить средним сроком жизненного цикла. Посчитаем его, выявив медиану разностей года старта и окончания продаж игр.

```python
# найдем средник срок жизненного цикла платформы
life_cicle_data = data.pivot_table(
    index = 'platform',
    values = 'year_of_release',
    aggfunc = ['min', 'max']
)
life_cicle_data['life_period'] = life_cicle_data['max'] - life_cicle_data['min'] + 1
f'Средний срок жизненного цикла платформы составляет {life_cicle_data["life_period"].median()} лет'
```

Как мы видим, каждая платформа получает достаточно стремительный рост, а потом - умеренное падение с возможными подъемами, которые не меняют общий стренд на снижение продаж. В этом и заключается жизненый цикл платформы: платформа выпускается на рынок, ее покупают для игр, а затем постпенно платформа устаревает, т.к. уже не может поддерживать актуальные игры. Важно обратить внимание на срок, с которым платформа проходит свои основные этапы жизненного цикла. В среднем за 2-3 года платформа достигает своего пика в продажах, а потом последние 4-5 лет продажи падают, что ведет к постепенному завершению жизненного цикла.
Исходя из вышеперечисленного, мы можем определить жизненный цикл платформы в 6-7 лет, что и является примерным сроком появления новых платформ, которые приходят на смену устаревшим.

```python
# Комментарий ревьюера
temp = data.copy() 
time_life = temp.pivot_table(index='platform', values='year_of_release', aggfunc=['min', 'max']).reset_index()
time_life['life'] = time_life['max'] - time_life['min'] + 1 # в срок жизни платформы, должны входить границы диапазона, 
                                                            # поэтому +1
time_life['life'].median()
```

### 3.4. Как оценки пользователей и критиков влияют на продажи игр?
Как мы уже выяснили, немало игр, которые дают хотя бы малейшие продажи, но редкая игра получает высокий спрос и продвигает платформу. Рассмотрим, как оценки пользователей и критиков влияют на продажи игры на платформе на примере PlayStation 4.

```python
# создадим диаграмму рассеивания к датафрейму
ps4_critic_score.plot(
    x = 'critic_score', 
    y = 'total_sales', 
    kind = 'scatter', 
    alpha = 0.7,
    title = 'Взаимосвязь оценок критиков и продаж игр на платформе PS4',
    grid = True,
    xlabel = 'Оценка критиков',
    ylabel = 'Продажи (млн)',
    color = 'b',
    s = 70
)
plt.show()
```

В случае с оценками критиков наблюдается более отчетливая взаимосвязь: чем выше продажи, тем сильнее разброс оценок критиков и тем больше пятна уходят вверх. Что касается корреляции, то мы наблюдаем среднюю корреляцию 0,4. Такой результат более красноречив, но так же показывет влияние оценки игры на продажи, хотя и не сильное.

Мы выявили хоть и слабую, но корреляцию оценок и продаж игр на определенной платформе, но справедливо ли такое утверждение в абсолютном выражении? Проверим эту же взаимосвязь на других платформах. Отличным вариантом будет взять для сравнения актуальные платформы. которые так же показывают хоршие продажи: 3DS, XBox One, WiiU, PS3 и XBox360.

```python
# создадим список интересующих нас платформ
less_profitable_platforms = ['3ds', 'xone', 'wiiu', 'ps3', 'x360']

# циклом построим диаграммы рассеяния и подсчитаем корреляцию
for platform_leader in less_profitable_platforms:
    # создание срезов оценок по одной платформе
    data_slice_user = data_years_filtered[
        (data_years_filtered['platform'] == platform_leader) & 
        (data_years_filtered['user_score'] != -1)
         ]
    data_slice_critic = data_years_filtered[
        (data_years_filtered['platform'] == platform_leader) & 
        (data_years_filtered['critic_score'] != -1)
         ]
    # визуализация оценок пользователей
    data_slice_user.plot(
        x = 'user_score',
        y = 'total_sales',
        kind = 'scatter',
        alpha = 0.7,
        grid = True,
        xlabel = 'Оценка критиков',
        ylabel = 'Продажи (млн)',
        color = 'b',
        s = 70,
        title = f'Взаимосвязь пользовательских оценок и продаж игр на платформе {platform_leader}'
    )
    plt.show()
    # визуализация оценок критиков
    data_slice_critic.plot(
        x = 'critic_score',
        y = 'total_sales',
        kind = 'scatter',
        alpha = 0.7,
        grid = True,
        xlabel = 'Оценка критиков',
        ylabel = 'Продажи (млн)',
        color = 'b',
        s = 70,
        title = f'Взаимосвязь оценок критиков и продаж игр на платформе {platform_leader}'
    )
    plt.show()
    print(f'Корреляция продаж игр для {platform_leader} и пользовательских оценок: {data_slice_user["total_sales"].corr(data_slice_user["user_score"])}')
    print(f'Корреляция продаж игр для {platform_leader} и оценок критиков: {data_slice_user["total_sales"].corr(data_slice_user["critic_score"])}')
    print('='*125)
```

Как мы можем наблюдать, зависимости достаточно сильно разнятся, и можно утверждать, что зависимость продаж игр на отдельных платформах и оценки критиков и пользователей платформозависимы. Но посмотрим отдельно каждый, держа в уме, что у PS4 была едва заметна взаимосвязь пользовательских оценок с продажами и составляла -0,04, а взаимосвязь оценок критиков с продажами была достаточно выражена с коэффициентом в 0,04:
- **3ds**: как у пользовательских оценок, так и у оценок критиков заметна слабовыраженная корреляция с коэффициентом около 0,2;
- **xone**: пользовательские оценки слабо демонстрируют тренд с коэффициентов -0,07, но оценки критиков достаточно хорошо отражают продажи, что подтверждается достаточно весомым коэффициентом в 0,37;
- **wiiu**: оба вида оценок демоснтируют умеренную взаимосвязь с продажами и насчитываеют коэффициент в 0,4 для пользователей и в 0,33 для критиков;
- **ps3**: здесь показатели достаточно сильно отличаются от предыдущих платформ, т.к. заметить здесь тренд весьма проблематично, а коэффициенты здесь показывают обратную корреляцию (около -0,16 для обоих видов оценок);
- **x360**: XBox 360 в целом повторяет разброс PS3, т.к. в диаграммах сложно выявить хоть какой-нибудь тренд, а коэффициенты либо незначительны, как у критиков (-0,5), либо показывают слабую обратную корреляцию, как у пользователей (-0,14).

Учитывая отмеченное выше, можно еще раз подтвердить платформонезависимость оценок пользователей и критиков, однако чаще встречается положительная корреляция оценок критиков с продажами, нежели взаимосвязь с пользовательскими оценками.

Помимо оценок важно учитывать и вкусовые предпочтения игроков и спрос на тот или иной тип геймплея, поэтому дополнительно рассмотрим то, как жанровые характеристики игр влияют на продажи.

## 5. Проверка гипотез
Полученные данные представляют ценность не только для исследовательского анализа с целью выявить общие тренды и закономерности игрового рынка, но и провести статистическое тестирование данных для успешной маркетинговой кампании на 2017 год. В данном отношении интерсует два вопроса:

1) Средние пользовательские рейтинги платформ Xbox One и PC одинаковые?
2) Средние пользовательские рейтинги жанров Action и Sports разные?

Для ответа на эти вопросы сформулируем гипотезы и проверим их на t-тестах, предварительно выбрав уровень стат. значимости в значении `alpha = 0,05`.

### 5.1. t-тест cредних пользовательских рейтингов платформ Xbox One и PC
Во-первых, если мы имеем данные пользовательских рейтингов платформ Xbox One и PC, будет полезно выяснить, равны ли они или же отличаются.

Сформулируем гипотезы:

***H0**: средний пользовательский рейтинг Xbox One = средний пользовательский рейтинг PC.*

***H1**: средний пользовательский рейтинг Xbox One != средний пользовательский рейтинг PC.*

Для проверки гипотезы воспользуемся t-тестом для двух отдельных выборок с двустронней альтернативной гипотезой. Для такого теста нам понадобится выборки генеральных совокупностей (ГС), однако сначала проверим дисперсии ГС, поскольку такой тест требует схожих дисперсий для коректного теста.

```python
# высчитаем дисперсию для ГС XBox
xbox_pop = data_years_filtered[(data_years_filtered['platform'] == 'xone') & 
                               (data_years_filtered['user_score'] >= 0)]['user_score']
np.var(xbox_pop)
```

```python
# высчитаем дисперсию для ГС PC
pc_pop = data_years_filtered[(data_years_filtered['platform'] == 'pc') & 
                             (data_years_filtered['user_score'] >= 0)]['user_score']
np.var(pc_pop)
```

Как мы видим, дисперсия ГС XBox меньше дисперсии ГС PC. Данное требование не критично - его можно невелировать посредством параметра `"equal_var"`.

```python
# проведем t-тест для двух отдельных выборок
test_result = st.ttest_ind(xbox_pop, pc_pop, equal_var=False)
test_result
```

Мы получили `p-value` в 0.11, что больше заданного уровня статистической значимости, что *не позволяет нам отвергнуть нулевую гипотезу в пользу альтернативной*. Поэтому нам следует *склониться к мнению (но не быть уверенными на 100%)*, что средние пользовательских оценок платформ Xbox One и PC равны.

В качестве дополнения отметим, что при уровне стат. значимости в 0,01 мы бы отвергнуть нулевую гипотезу не смогли.

```python
print(f'Средний пользовательский рейтинг жанра Xbox One: {xbox_pop.mean()}')
print(f'Средний пользовательский рейтинг жанра PC: {pc_pop.mean()}')
```

Средние значения жанров близки в пределах единицы, что хорошо подтверждает верность наших результатов проверки гипотезы.

### 5.2. t-тест cредних пользовательских рейтингов жанров Action и Sports
Также интеренет-магазину важно учитывать удовлетворенность игроков от жанров. Попробуем проверить, отличаются ли средние пользовательские рейтинги жанров Action и Sports.

Сформулируем гипотезы:

***H0**: средний пользовательский рейтинг жанра Action = средний пользовательский рейтинг жанра Sports.*

***H1**: средний пользовательский рейтинг жанра Action != средний пользовательский рейтинг жанра Sports.*

Для проверки гипотезы воспользуемся t-тестом для двух отдельных выборок с двустронней альтернативной гипотезой. Для такого теста нам понадобится выборки генеральных совокупностей (ГС), однако сначала проверим дисперсии ГС, поскольку такой тест требует схожих дисперсий для коректного теста.

```python
# высчитаем дисперсию для ГС action
action_pop = data_years_filtered[(data_years_filtered['genre'] == 'action') & 
                                 (data_years_filtered['user_score'] >= 0)]['user_score']
np.var(action_pop)
```

```python
# высчитаем дисперсию для ГС sports
sports_pop = data_years_filtered[(data_years_filtered['genre'] == 'sports') & 
                                 (data_years_filtered['user_score'] >= 0)]['user_score']
np.var(sports_pop)
```

Как мы видим, дисперсия ГС action меньше дисперсии ГС sports. Тогда укажем парметр "equal_var" как ложный.

```python
# создадим выборки для двух ГС
# action_sample = action_pop.sample(30)
# sports_sample = sports_pop.sample(30)

# проведем t-тест для двух отдельных выборок
test_result = st.ttest_ind(action_pop, sports_pop, equal_var=False)
test_result
```

Мы получили крайне низкое `p-value` в 2.87e-19, что представляет собой крайне низкий показатель и *нам следует отвергнуть нулевую гипотезу*, поэтому у нас есть основания пологать, что средний пользовательский рейтинг жанра Action не равен среднему пользовательскому рейтингу жанра Sports.

Подкрепим проверенные тесты средними пользовательскими рейтингами по платформам и жанрам.

```python
print(f'Средний пользовательский рейтинг жанра action: {action_pop.mean()}')
print(f'Средний пользовательский рейтинг жанра sports: {sports_pop.mean()}')
```
